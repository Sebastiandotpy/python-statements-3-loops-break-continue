```python
# Task 1: Omitting the number

def print_numbers_without_zero():
    """
    Print numbers from 0 to 8, omitting 3 and 6.
    """
    numbers = [num for num in range(9) if num not in (3, 6)]
    print(*numbers)


# Task 2: Breaking the inner loop

def print_nested_numbers():
    """
    Print numbers in three separate lines:
    - First line: numbers from 0 to 7
    - Second line: numbers from 10 to 17
    - Third line: numbers from 20 to 27
    """
    for i in range(3):
        line = ""
        for j in range(8):
            line += str(i * 10 + j) + " "
        print(line)
    print()


# Task 3: Break or continue

def process_numbers():
    """
    Prompt the user for two integers and process them according to the rules:
    - If the first number is smaller than the second number, find numbers divisible by 3 between them.
    - If the first number is larger than the second number, find the first number divisible by 5 between them.
    - If the numbers are equal, print that information.
    """
    first_number = int(input("First number: "))
    second_number = int(input("Second number: "))

    if first_number < second_number:
        for num in range(first_number, second_number + 1):
            if num % 3 == 0:
                print(num, "is divisible by 3.")
        if first_number % 3 != 0:
            print("No numbers divisible by 3 between the given range.")
    elif first_number > second_number:
        for num in range(first_number, second_number - 1, -1):
            if num % 5 == 0:
                print(num, "is divisible by 5.")
                break
        if first_number % 5 != 0:
            print("No number divisible by 5 between the given range.")
    else:
        print("Both numbers are equal.")

    print()


# Task 4: Divisor game

def play_divisor_game():
    """
    Prompt the user for two integers and play the divisor game:
    - The first integer must be bigger than 100 and must be odd.
    - The second integer must be odd.
    - If the first number is divisible by the second number, earn a point and continue.
    - If the first number is not divisible by the second number, quit the game.
    - Print the number of points earned.
    """
    points = 0
    first_number = int(input("First number: "))
    second_number = int(input("Second number: "))

    while first_number > 100 and first_number % 2 != 0 and second_number % 2 != 0:
        if first_number % second_number == 0:
            points += 1
            print("You earned a point!")
        else:
            print("You've made a mistake!")
            break

        first_number = int(input("First number: "))
        second_number = int(input("Second number: "))

    print("\nYou gathered", points, "point(s)!")


# Task 5: Only digits

def count_digits():
    """
    Prompt the user to enter characters and count the sum of the digits in the string.
    Terminate the program after encountering the '=' character.
    """
    characters = input("Input your characters: ")
    digit_sum = 0

    for char in characters:
        if char == '=':
            break
        if char.isdigit():
            digit = int(char)
            digit_sum += digit
            print("Found a digit", digit)

    print("\nSum of digits:", digit_sum)

print("Task 1:")
print_numbers_without_zero()

print("Task 2:")
print_nested_numbers()

print("Task 3:")
process_numbers()

print("Task 4:")
play_divisor_game()

print("Task 5:")
count_digits()
```
